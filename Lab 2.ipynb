```python
from transformers import pipeline, set_seed
```

```python
generator = pipeline("text-generation", model="gpt2")
set_seed(42)  # For reproducibility
```

```python
prompt = "Generative AI will revolutionize education because"
output_default = generator(prompt, max_length=50, num_return_sequences=1)
print("Default Output:\n", output_default[0]['generated_text'][:150], "...")
```

```python
for temp in [0.3, 0.7, 1.0, 1.5]:
    print(f"\n--- Temperature = {temp} ---")
    output = generator(prompt, max_length=50, temperature=temp, num_return_sequences=1)
    print(output[0]['generated_text'][:150], "...")
```

```python
for k in [10, 50, 100]:
    print(f"\n--- Top-k = {k} ---")
    output = generator(prompt, max_length=50, top_k=k, num_return_sequences=1)
    print(output[0]['generated_text'][:150], "...")
```

```python
for p in [0.7, 0.9, 0.95]:
    print(f"\n--- Top-p = {p} ---")
    output = generator(prompt, max_length=50, top_p=p, num_return_sequences=1)
    print(output[0]['generated_text'][:150], "...")
```

```python
for rp in [1.0, 1.5, 2.0]:
    print(f"\n--- Repetition Penalty = {rp} ---")
    output = generator(prompt, max_length=60, repetition_penalty=rp, num_return_sequences=1)
    print(output[0]['generated_text'][:150], "...")
```

```python
custom_output = generator(
    prompt,
    max_length=60,
    temperature=0.8,
    top_k=50,
    top_p=0.9,
    repetition_penalty=1.2,
    num_return_sequences=2
)

for i, out in enumerate(custom_output):
    print(f"\nCustom Output {i+1}:\n", out['generated_text'][:150], "...")
```
